```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "nba_hurt.csv"
inj_df = pd.read_csv(file)
```

```python
inj_uni_df = inj_df.groupby(["college"])
inj_uni_count = inj_uni_df["college"].count()
inj_uni = inj_uni_df["college"].unique()
inj_uni_dfa = pd.DataFrame({"Colleges":inj_uni,
                            "Count": inj_uni_count})
inj_uni_dfa.head()
inj_uni_dfb = inj_uni_dfa.sort_values(["Count"], ascending = False)
inj_uni_dfb
```
