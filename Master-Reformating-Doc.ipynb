```python
import pandas as pd
import numpy as np
```

```python
file = "nba_hurt.csv"
nba_hurt = pd.read_csv(file)
nba_hurt.head()
```

```python
#splitting height feet and inches to convert each to meters, then later adding them together
# new data frame with split value columns 
#new = data["Name"].str.split(" ", n = 1, expand = True)   
# making separate first name column from new data frame 
#data["First Name"]= new[0]   
# making separate last name column from new data frame 
#data["Last Name"]= new[1]  
# Dropping old Name columns 
#data.drop(columns =["Name"], inplace = True)   
# df display 
#data 

heightsplit=nba_hurt["height"].str.split("-", n=1, expand=True)
nba_hurt["feet"]=heightsplit[0]
nba_hurt["inches"]=heightsplit[1]
nba_hurt.drop(columns=["height"], inplace=True)
nba_hurt.head()
```

```python
#convert feet to integer, convert feet to inches
nba_hurt['feets']=(nba_hurt['feet'].apply(int)*12)+nba_hurt["inches"].apply(int)
#nbadata_df=nbadata_df["feet"].mul(12)
nba_hurt.head()
```

```python
#nbadata_df[["feets", "weight","Notes"]].head()

nba_hurt['kilos']=nba_hurt['weight'].divide(2.2)
#nbadata_df.head()
#converting weight to kilograms
#weightmetric=nbadata_df["weight"].divide(2.2)
#weightmetric.head()        

#metricheight=(nbadata_df["feets"].divide(39.97)), (nbadata_df["feets"].multiply["feets"])
#metricheight=(nbadata_df["feets"].divide(39.97)), (nbadata_df["feets"].sqr)
nba_hurt["metricheight"]=(nba_hurt["feets"].divide(39.97)) **2
#, (nbadata_df["feets"]**2)
nba_hurt.head()
```

```python
##Adding 'top four' injury types to data set.
conditions = nba_hurt["Strains/Sprains"] = nba_hurt["Notes"].str.contains("strained") | nba_hurt["Notes"].str.contains("sprained")
nba_hurt["Dislocations"] = nba_hurt["Notes"].str.contains("dislocated")
nba_hurt["Arm Fractures"] = nba_hurt["Notes"].str.contains("fractured") & (nba_hurt["Notes"].str.contains("wrist") | nba_hurt["Notes"].str.contains("finger")| nba_hurt["Notes"].str.contains("arm")| nba_hurt["Notes"].str.contains("elbow"))
nba_hurt["Leg Fractures"] = nba_hurt["Notes"].str.contains("fractured") & (nba_hurt["Notes"].str.contains("wrist") | nba_hurt["Notes"].str.contains("ankle")| nba_hurt["Notes"].str.contains("leg")| nba_hurt["Notes"].str.contains("shin"))
conditions = [
    (nba_hurt['Strains/Sprains'] == True),
    (nba_hurt['Dislocations'] == True),
    (nba_hurt["Arm Fractures"] == True),
    (nba_hurt["Leg Fractures"] == True)
]
choices = ['Strain/Sprain', 'Dislocation', 'Arm Fracture', 'Leg Fracture']

nba_hurt["Injury Type"] = np.select(conditions, choices, default= 'No Index')
```

```python
## Add average medical costs 
#Rough index based on report: https://aspe.hhs.gov/report/common-sports-injuries-incidence-and-average-charges
conditions2 = [
    (nba_hurt['Injury Type'] == "Strain/Sprain"),
    (nba_hurt['Injury Type'] == "Dislocation"),
    (nba_hurt["Injury Type"] == "Arm Fracture"),
    (nba_hurt["Injury Type"] == "Leg Fracture")
]
choices2 = [3175, 4575, 7666, 3403]

nba_hurt["Estimated Medical Costs (2012)"] = np.select(conditions2, choices2, default= 'No Index')
```

```python
##At this point, we saved this document as Nba_hurt2.csv, 
#but we used MicrosoftExcel to change the format of birth_date, so know that that's going on
#in the background
nba_hurt.to_csv("nba_hurt3.csv")
```
