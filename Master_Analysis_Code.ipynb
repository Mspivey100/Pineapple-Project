```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
##Bring in data and drop any NAs
nba_hurt = pd.read_csv('nba_hurt5.csv')
nba_hurt.head()
nba_hurt = nba_hurt.drop([0], axis = 0)
nba_hurt.dropna(how = "any")
```

```python
##Make graphs on ages
inj_age_df2 = nba_hurt.groupby(["Injury Age"])
inj_age_count2 = inj_age_df2["Injury Age"].count()
inj_age_df = nba_hurt.groupby(["Injury Age"])
inj_age_count = inj_age_df["Injury Age"].count()
inj_age_count.head()
```

```python
x_axis2 = np.arange(0,len(inj_age_count2),1)
plt.bar(x_axis2,inj_age_count2, label="Age of Player ", align = "center")
plt.xticks([0,3,7,11,15,19,23,27,31,35],["19","22","26","30","34","38","42","46","50","54"] )
plt.xlabel("Player Age at time of injury")
plt.ylabel("Number of Injuries")
```

```python
##Make a graph by team (note, couldn't put names down)
inj_team = nba_hurt.groupby(["Team"])
inj_team_count = inj_team["name"].count()
inj_team_df = pd.DataFrame({"Number of Injuries": inj_team_count})
team_df = inj_team_df.sort_values("Number of Injuries", ascending = False)
```

```python
teams = nba_hurt["Team"].unique()
x_axis = np.arange(len(teams))
plt.bar(x_axis,inj_team_count, label = "Injuries")
plt.xlabel("Teams")
plt.ylabel("Injuries")
```

```python
##Make graph for universities/ colleges
inj_uni_df = nba_hurt.groupby(["college"])
inj_uni_count = inj_uni_df["college"].count()
inj_uni = inj_uni_df["college"].unique()
inj_uni_dfa = pd.DataFrame({"Colleges":inj_uni,
                            "Count": inj_uni_count})
inj_uni_dfa.head()
inj_uni_dfb = inj_uni_dfa.sort_values(["Count"], ascending = False)
inj_uni_dfb.head()
```

```python
%matplotlib notebook
x_axis = np.arange(len(inj_uni))
plt.bar(x_axis, inj_uni_count, label = "Colleges")
plt.xlabel("Colleges")
plt.ylabel("Injuries")
```
