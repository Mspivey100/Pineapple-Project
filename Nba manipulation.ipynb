```python
import pandas as pd
from pprint import pprint
```

```python
file='nba_hurt.csv'
nba_df=pd.read_csv(file)
nba_df.head()
```

```python
#injury by position
injury_by_position=nba_df.groupby(['position'])
injury_p_count=injury_by_position['name'].count()
injury_Pcount_df=pd.DataFrame({'Injuries':injury_p_count})
injury_Pcount_df
```

```python
#injury by college
injury_by_college=nba_df.groupby(['college'])
injury_c_count=injury_by_college['name'].count()
injury_Ccount_df=pd.DataFrame({'Injuries':injury_c_count})
injury_Ccount_df.sort_values(by='Injuries',ascending=False).head()
```

```python
#injury by draft year
injury_by_draftyear=nba_df.groupby(['year_start'])
injury_dy_count=injury_by_draftyear['name'].count()
injury_DYcount_df=pd.DataFrame({'Injuries':injury_dy_count})
injury_DYcount_df.sort_values(by='Injuries',ascending=False).head()
```

```python
#player with the most injuries
#david lee shoud not be included
injury_by_name=nba_df.groupby(['name'])
injury_name_count=injury_by_name['name'].count()
injury_Ncount_df=pd.DataFrame({'Injuries':injury_name_count})
injury_Ncount_df.sort_values(by='Injuries',ascending=False).head()
```
