```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
inj_file = "nba_hurt.csv"
inj_df = pd.read_csv(inj_file)
```

```python
inj_df.head()
```

```python
inj_pos_df = inj_df.groupby(["position"])
inj_pos_count = inj_pos_df["position"].count()
inj_pos = inj_pos_df["position"].unique()
inj_pos_dfa = pd.DataFrame({"Position":inj_pos, 
                            "Count": inj_pos_count})
inj_pos_dfa.head
```
