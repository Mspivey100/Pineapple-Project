```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
inj_file = "nba_hurt2.csv"
inj_df = pd.read_csv(inj_file)
```

```python
inj_df.head()
```

```python
inj_pos_df = inj_df.groupby(["position"])
inj_pos_count = inj_pos_df["position"].count()
inj_pos = inj_pos_df["position"].unique()
inj_pos_dfa = pd.DataFrame({"Position":inj_pos, 
                            "Count": inj_pos_count})
inj_pos_dfa
```

```python
positions = inj_pos_dfa["Position"]
positions
```

```python
%matplotlib notebook
x_axis = np.arange(len(positions))
plt.bar(x_axis, inj_pos_count, align ="center")

plt.xticks(x_axis,["C", "C-F","F", "F-C","F-G", "G", "G-F"])
plt.xlabel("Positions")
plt.ylabel("Number of Injuries")
plt.savefig("Injuries Position.png")
```
