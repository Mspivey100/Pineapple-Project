```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats 
```

```python
csv_file = "nba_hurt.csv"
```

```python
inj_df = pd.read_csv(csv_file)
inj_df.head()
```

```python
inj_team = inj_df.groupby(["Team"])
inj_team_count = inj_team["name"].count()
inj_team_df = pd.DataFrame({"Number of Injuries": inj_team_count})
team_df = inj_team_df.sort_values("Number of Injuries", ascending = False)
team_df
```

```python
team_df.describe()
```

```python
team_df.head(5)
teams = inj_df["Team"].unique()
x_axis = np.arange(len(teams))
```

```python
plt.bar(x_axis,inj_team_count, label = "Injuries")
plt.xlabel("Teams")
plt.ylabel("Injuries")
```
