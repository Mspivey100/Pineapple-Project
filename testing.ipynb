```python
import pandas as pd
import json
from confiy import api_key
import requests
from pprint import pprint
url = 'https://api.sportradar.us/nba/trial/v5/en/league/injuries.json?api_key='
request_url = url + api_key
response = requests.get(request_url)
print(response.status_code)
```

```python
response_json=response.json()
pprint(response_json)
#for each team in teams loop through 
#for play in players of teams 
#get full name and injury 

        


    
    
```

```python
Nba_hurt=[]
teams=response_json['teams']
for team in teams:
    team_name= team['name']
    player_id=team.get('players')
    for player in player_id:
        Nba_hurt.append({
            'Player':player['full_name'],
            'Team':team_name,
            'Position':player['position'],
            'Injury info':player['injuries'],
            'Player_ID':player['id']
        })
```
